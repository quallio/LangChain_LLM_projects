```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

### Comparing embeddings

```python
query_result_1 = embeddings.embed_query("I have a car")
query_result_2 = embeddings.embed_query("I like cars")
query_result_3 = embeddings.embed_query("The parrot is green")
```

```python
import numpy as np
```

```python
np.dot(query_result_1, query_result_2)
```

```python
np.dot(query_result_2, query_result_3)
```

```python
np.dot(query_result_1, query_result_3)
```
